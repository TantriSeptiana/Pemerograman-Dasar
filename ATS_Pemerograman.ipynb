```python
# SOAL NOMOR 2

def hitung_efisiensi(input_energi, output_energi):
    efisiensi = output_energi / input_energi
    return efisiensi

# Pemanggilan fungsi
hasil = hitung_efisiensi(500, 400)

print(f"Efisiensi = {hasil}")             # Bentuk desimal
print(f"Efisiensi = {hasil * 100}%")      # Bentuk persen
```

Program digunakan untuk menghitung efisiensi turbin dengan cara membandingkan energi keluaran terhadap energi masukan. Program ini menggunakan fungsi bernama hitung_efisiensi() yang menerima dua nilai, yaitu input_energi dan output_energi, lalu menghitung nilai efisiensi dengan rumus output_energi / input_energi.

Baris penting pada program ini terletak pada bagian efisiensi = output_energi / input_energi, yang berfungsi untuk menghitung seberapa besar energi masuk yang berhasil diubah menjadi energi keluar. Nilai kemudian dikembalikan melalui perintah return efisiensi, agar bisa digunakan di luar fungsi. Lalu hasilnya ditampilkan dalam bentuk desimal dan juga persen melalui print(f"Efisiensi = {hasil * 100:.2f}%"). Program ini membantu mengevaluasi performa sistem dengan memberikan informasi efisiensi dalam bentuk yang mudah dipahami.

```python
# SOAL NOMOR 4

tekanan_uap = float(input("Masukkan tekanan uap (psi): "))
status_manual_override = bool(int(input("Masukkan status manual override (0=False, 1=True): ")))

if tekanan_uap > 150 and status_manual_override == False:
    print("Katup Terbuka")
else:
    print("Katup Tertutup")
```

Program menggunakan logika if majemuk untuk menentukan apakah katup pengaman boleh terbuka atau tidak. Program ini menerima dua input, yaitu nilai tekanan_uap dan status_manual_override. Katup hanya akan terbuka jika tekanan uap lebih besar dari 150 psi dan status manual override bernilai False.

Baris penting pada program ini adalah if tekanan_uap > 150 and status_manual_override == False:, yang menjadi bagian dari pengambilan keputusan. Operator logika and memastikan kedua kondisi tersebut harus benar agar perintah pada blok if dijalankan dan menampilkan “Katup Terbuka”. Jika salah satu kondisi tidak terpenuhi, maka blok else akan dijalankan dan program menampilkan “Katup Tertutup”. Program ini menyelesaikan masalah pengendalian keamanan pada sistem tekanan otomatis agar katup hanya aktif saat kondisi tertentu terpenuhi.

```python
# SOAL NOMOR 6

# Diketahui
tegangan = 220  # Volt
arus = 15       # Ampere

# Rumus daya listrik
daya = tegangan * arus

print("Tegangan (V) =", tegangan, "Volt")
print("Arus (I) =", arus, "Ampere")
print("Daya listrik (P) =", daya, "Watt")
```

Program ini digunakan untuk menghitung daya listrik berdasarkan nilai tegangan dan arus yang sudah diketahui. Program menetapkan tegangan sebesar 220 Volt dan arus sebesar 15 Ampere, lalu menghitung daya menggunakan rumus P = V * I.

Bagian penting program ini adalah daya = tegangan * arus, karena di situ proses perhitungan utama terjadi. Hasilnya kemudian ditampilkan dengan fungsi print() agar pengguna bisa melihat besar tegangan, arus, dan daya secara lengkap. Program ini membantu menghitung daya listrik dengan cara yang sederhana dan cepat, tanpa perlu melakukan perhitungan manual.

```python
# SOAL NOMOR 8

# Fungsi untuk menghitung efisiensi termal turbin gas
def hitung_efisiensi_termal(energi_keluar, energi_masuk):
    efisiensi = (energi_keluar / energi_masuk) * 100
    print("Energi keluar =", energi_keluar, "Joule")
    print("Energi masuk  =", energi_masuk, "Joule")
    print("Efisiensi termal =", efisiensi, "%")

# Pemanggilan fungsi
hitung_efisiensi_termal(5000, 10000)
```

Program ini digunakan untuk menghitung efisiensi termal pada turbin gas. Program mendefinisikan fungsi bernama hitung_efisiensi_termal() yang menerima dua nilai, yaitu energi keluar dan energi masuk. Pada fungsi ini, efisiensi dihitung dengan rumus (energi_keluar / energi_masuk) * 100, lalu hasilnya ditampilkan dalam bentuk persen.

Baris rumus perhitungan menjadi bagian penting karena menunjukkan seberapa besar energi yang bisa diubah menjadi kerja atau output. Ketika fungsi dipanggil dengan nilai 5000 Joule sebagai energi keluar dan 10000 Joule sebagai energi masuk, program akan menampilkan efisiensi termal sebesar 50%. Program ini membantu mempermudah perhitungan efisiensi sistem turbin secara cepat dan jelas tanpa perlu menghitung manual.

```python
# SOAL NOMOR 10

kapasitas_tangki = 100
isi_tangki = 0

while isi_tangki <= kapasitas_tangki:
  print (f"Tangki terisi : {isi_tangki} liter")

  if isi_tangki == 80:
    print("WARNING : Hampir penuh!")
  elif isi_tangki == 100:
    print("TANGKI PENUH - Pengisian berhenti!")
    break
  isi_tangki += 10
```

```python
# CHALLENGE

# Standar kualitas air boiler
standar_boiler = {
    "Air Umpan": {
        "pH": (8.5, 9.5),
        "TDS": (0, 100),
        "Konduktivitas": (0, 100),
        "Oksigen": (0, 0.05)
    },
    "Air Drum": {
        "pH": (10, 11),
        "TDS": (2000, 3500),
        "Konduktivitas": (3000, 6000),
        "Oksigen": (0, 0.05)
    }
}

def cek_parameter(nama, nilai, batas):
    if nilai < batas[0]:
        return f"{nama} terlalu rendah ({nilai})"
    elif nilai > batas[1]:
        return f"{nama} terlalu tinggi ({nilai})"
    else:
        return f"{nama} sesuai standar ({nilai})"

def analisis_air(jenis, data):
    print(f"\nAnalisis Kualitas {jenis}")
    batas = standar_boiler[jenis]
    for parameter in data:
        hasil = cek_parameter(parameter, data[parameter], batas[parameter])
        print(hasil)

# hasil data dari sensor
data_air_umpan = {
    "pH": 9.0,
    "TDS": 85,
    "Konduktivitas": 95,
    "Oksigen": 0.04
}

data_air_drum = {
    "pH": 10.5,
    "TDS": 3600,
    "Konduktivitas": 6200,
    "Oksigen": 0.06
}

# Loop while untuk analisis berulang
jenis_air = ["Air Umpan", "Air Drum"]
index = 0

while index < len(jenis_air):
    if jenis_air[index] == "Air Umpan":
        analisis_air("Air Umpan", data_air_umpan)
    elif jenis_air[index] == "Air Drum":
        analisis_air("Air Drum", data_air_drum)
    else:
        print("Jenis air tidak dikenali.")
    index += 1
```

Jadi untuk challenge kami membuat analisis kualitas air boiler. Karena kualitas air boiler dapat mempengaruhi umur pakai, efisiensi, dan keamanan operasi boiler. Dan menurut kami ini adalah sebuah masalah di bidang energi, karena berkaitan dengan efisiensi dan keamanan. Inspirasi kami berawal dari mata kuliah Kimia Dasar

Program ini dirancang untuk menganalisis kualitas air pada sistem boiler, khususnya pada dua jenis air yaitu, Air Umpan Boiler dan Air Drum Boiler. Program akan mengevaluasi apakah parameter-parameter penting seperti pH, TDS, konduktivitas, dan oksigen terlarut berada dalam rentang standar yang telah ditentukan berdasarkan tabel kualitas air boiler. Karena air boiler yang buruk memiliki banyak dampak antara lain :

1. Scaling (kerak)
 Akibat pengendapan garam CaCO₃, Mg(OH)₂, CaSO₄.
 Menghambat perpindahan panas → efisiensi turun.
 Menyebabkan overheating dan kerusakan pipa boiler.
2. Korosi
 Disebabkan oleh pH rendah, O₂ terlarut, atau CO₂ dalam air.
 Mengurangi umur pakai pipa dan drum boiler.
3. Foaming & Carryover
 TDS terlalu tinggi → terbentuk busa (foam) di permukaan air.
 Dapat terbawa ke jalur uap → menurunkan kualitas uap.

Untuk bagian standar_boiler kami menggunakan dictionary nested untuk menyimpan data batas atas dan bawah untuk parameter kualitas air.

Fungsi cek_parameter ()
Fungsi ini menerima nama, nilai batas dan fungsinya juga menggunakan logika untuk menentukan apakah nilai terlalu rendah.

Fungsi analisis_air()
Fungsi ini menerima, jenis, data.
Fungsi ini Mengambil standar dari standar_boiler. Menggunakan loop for untuk mengevaluasi setiap parameter. Memanggil cek_parameter() untuk setiap parameter.

```python
import pandas as pd

# Membuat data dummy mahasiswa
data = {
    'Nama': ['Aldi', 'Budi', 'Citra', 'Dewi', 'Eka', 'Fajar'],
    'Umur': [20, 21, 22, None, 23, 21],
    'Nilai_Tugas': [85, 90, None, 78, 88, 90],
    'Nilai_UAS': [80, 92, 85, 70, None, 95],
    'Jurusan': ['Energi', 'Energi', 'Teknik', 'Teknik', 'Energi', None]
}

df = pd.DataFrame(data)
df
```

Program dimulai dengan mengimpor library pandas, yang digunakan untuk mengolah data dalam bentuk tabel. Setelah itu dibuat sebuah variabel bernama data yang berisi beberapa daftar seperti yang ditampilkan. Setiap daftar berisi beberapa nilai, dan sebagiannya sengaja dibiarkan kosong (None) untuk menunjukkan contoh data yang tidak lengkap.

Setelah data disusun, data akan diubah menjadi sebuah tabel menggunakan perintah df = pd.DataFrame(data) sehingga tabel akan menampilkan semua isi datanya.

```python
df.info()
```

Perintah df.info() digunakan untuk menampilkan informasi mengenai struktur tabel, seperti jumlah baris dan kolom, tipe data setiap kolom, dan jumlah nilai yang kosong.

```python
df.describe()
```

Perintah df.describe() akan menampilkan ringkasan statistik untuk kolom yang berisi angka, misalnya rata-rata, nilai minimum, maksimum, dan nilai lainnya, sehingga kita bisa memahami gambaran umum dari data numeriknya.

```python
df.isnull().sum()
```

Perintah df.isnull().sum() digunakan untuk menghitung berapa banyak nilai yang kosong pada masing-masing kolom. Dari hasilnya terlihat ada beberapa kolom yang memiliki satu nilai kosong.

```python
df.duplicated().sum()
```

Perintah df.duplicated().sum() berfungsi untuk memeriksa apakah ada baris yang duplikat di dalam tabel. Hasilnya menunjukkan tidak ada data yang berulang.

```python
# Mengisi nilai kosong di kolom numerik dengan nilai rata-rata
df['Umur'] = df['Umur'].fillna(df['Umur'].mean())
df['Nilai_Tugas'] = df['Nilai_Tugas'].fillna(df['Nilai_Tugas'].mean())
df['Nilai_UAS'] = df['Nilai_UAS'].fillna(df['Nilai_UAS'].mean())

# Mengisi nilai kosong di kolom kategorik dengan 'Tidak Diketahui'
df['Jurusan'] = df['Jurusan'].fillna('Tidak Diketahui')

df
```

Program melakukan pengisian nilai kosong pada kolom yang berisi angka. Setiap kolom angka yang kosong diisi menggunakan nilai rata-rata dari kolom masing-masing.

Setelah semua kolom angka diisi, program melanjutkan pengisian ke kolom Jurusan. Karena kolom ini berisi teks, program tidak bisa menggunakan rata-rata. Oleh karena itu, nilai kosong diganti dengan tulisan Tidak Diketahui.

```python
# Tambahkan satu data duplikat untuk simulasi
df.loc[len(df)] = ['Aldi', 20, 85, 80, 'Energi']

df
```

Program melakukan duplikasi terhadap data mahasiswa, sehingga semua data dengan nama "Aldi" berulang sebanyak 3 baris.

```python
# Cek jumlah duplikat
print("Jumlah duplikat sebelum dihapus:", df.duplicated().sum())

df
```

Program digunakan untuk memeriksa duplikasi pada data mahasiswa, perintah df.duplicated() akan mengecek setiap baris dan menandai baris yang sama persis dengan baris sebelumnya sebagai duplikat. Dari hasil program tersebut terdapat 1 data yang mengalami duplikat.

```python
# Hapus duplikat
df = df.drop_duplicates()
print("Jumlah duplikat setelah dihapus:", df.duplicated().sum())
```

Program digunakan untuk menghapus data yang berulang/duplikat, sehingga pada output jumlah duplikat yang ditampilkan menjadi 0.

```python
# Ubah semua nama jurusan menjadi huruf kapital
df['Jurusan'] = df['Jurusan'].str.upper()

df
```

Program berfungsi untuk mengubah semua nama jurusan menggunakan huruf kapital. Perintah .str.upper() mengubah seluruh teks di kolom Jurusan menjadi huruf kapital.

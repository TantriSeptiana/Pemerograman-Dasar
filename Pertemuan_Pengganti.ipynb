```python
# SOAL 3: Menghitung Rata-rata dari List Nilai
def hitung_rata_rata(data_nilai):
    """Menghitung rata-rata dari list angka menggunakan perulangan for."""
    total = 0

    # Perulangan untuk menjumlahkan nilai
    for nilai in data_nilai:
        total += nilai

    # Pembagian untuk rata-rata
    if total > 0:
        return total / 5
    return 0

# Pemanggilan:
nilai_praktikan = [85, 78, 92, 65, 90]
print(f"Nilai rata-rata kelas adalah: {hitung_rata_rata(nilai_praktikan)}")
```

```python
#SOAL 3 MODIFIKASI
def hitung_rata_rata(data_nilai):
    """Menghitung rata-rata nilai dengan penalti jika ada nilai tidak valid (di bawah 0 atau di atas 100)."""
    total = 0
    jumlah_valid = 0
    penalti = 0  # penalti 5 poin per nilai tidak valid

    for nilai in data_nilai:
        if nilai < 0 or nilai > 100:
            print(f"Nilai {nilai} tidak valid! Menambahkan penalti 5 poin.")
            penalti += 5
        else:
            total += nilai
            jumlah_valid += 1

    # Jika ada nilai valid, hitung rata-rata
    if jumlah_valid > 0:
        rata_rata = total / jumlah_valid
        # Kurangi rata-rata dengan penalti (tidak boleh negatif)
        rata_rata_final = max(rata_rata - penalti, 0)
        return rata_rata_final
    else:
        return 0

# Pemanggilan:
nilai_praktikan = [85, 78, 105, -10, 90]
print(f"Nilai rata-rata akhir: {hitung_rata_rata(nilai_praktikan)}")
```

```python
#SOAL 4 MODIFIKASI
def tentukan_predikat(skor):
    """Menentukan predikat (A, B, C, D, E) berdasarkan skor dengan validasi tipe data."""
    # Pengecekan tipe data
    if not isinstance(skor, (int, float)):
        return "Error: Input harus berupa angka!"

    # Pengecekan rentang nilai
    if skor < 0 or skor > 100:
        return "Error: Skor harus antara 0 - 100!"

    # Penentuan predikat
    if skor >= 85:
        return "A"
    elif skor >= 70:
        return "B"
    elif skor >= 60:
        return "C"
    elif skor >= 50:
        return "D"
    else:
        return "E"

# Pemanggilan:
print(f"Skor 92 mendapat predikat: {tentukan_predikat(92)}")
print(f"Skor 67 mendapat predikat: {tentukan_predikat(67)}")
print(f"Skor 'bagus' mendapat predikat: {tentukan_predikat('bagus')}")
print(f"Skor 120 mendapat predikat: {tentukan_predikat(120)}")
```

```python
# SOAL 5: Menghitung Faktorial dengan Loop While
def hitung_faktorial(n):
    """Menghitung faktorial dari bilangan positif n."""
    hasil = 1

    # Perulangan berjalan selama n lebih besar dari 1
    while n > 1:
        hasil *= n
        n -= 1  # Kurangi n
    return hasil

# Pemanggilan:
bilangan_n = 4
print(f"Faktorial dari {bilangan_n} adalah: {hitung_faktorial(bilangan_n)}")  # Output: 24
```

```python
# SOAL 5 MODIFIKASI
def hitung_faktorial(n):
    """Menghitung faktorial dari bilangan n dengan pengecekan kondisi khusus."""

    if n < 0:
        return "Faktorial tidak terdefinisi untuk bilangan negatif."
    elif n == 0:
        return 1  # Faktorial dari 0 adalah 1
    else:
        hasil = 1
        while n > 1:
            hasil *= n
            n -= 1
        return hasil

# Pemanggilan:
bilangan_n = int(input("Masukkan bilangan untuk dihitung faktorialnya: "))
print(f"Faktorial dari {bilangan_n} adalah: {hitung_faktorial(bilangan_n)}")
```

```python
# SOAL 6: Analisis Data Suhu Ekstrem (for dan if/elif)
def analisis_suhu(data_suhu):
    """Menghitung jumlah hari Panas dan Dingin dari data_suhu."""
    panas_count = 0
    dingin_count = 0

    for suhu in data_suhu:
        if suhu >= 35:
            panas_count += 1
        elif suhu <= 15:
            dingin_count += 1

    return panas_count, dingin_count

# Pemanggilan:
suhu_mingguan = [36, 30, 14, 25, 38, 16, 28]
jumlah_panas, jumlah_dingin = analisis_suhu(suhu_mingguan)
print(f"Jumlah hari Panas (>=35°C): {jumlah_panas}")
print(f"Jumlah hari Dingin (<=15°C): {jumlah_dingin}")
```

```python
# SOAL 6 MODIFIKASI
def analisis_suhu(data_suhu, batas_panas=35, batas_dingin=15):
    """
    Menghitung jumlah hari Panas, Dingin, dan Normal dari data suhu.
    Parameter:
      data_suhu     : list berisi data suhu harian
      batas_panas   : suhu minimum agar dikategorikan Panas (default 35)
      batas_dingin  : suhu maksimum agar dikategorikan Dingin (default 15)
    """
    hasil = {
        "Panas": 0,
        "Dingin": 0,
        "Normal": 0
    }

    for suhu in data_suhu:
        if suhu >= batas_panas:
            hasil["Panas"] += 1
        elif suhu <= batas_dingin:
            hasil["Dingin"] += 1
        else:
            hasil["Normal"] += 1

    return hasil

# --- Pemanggilan ---
suhu_mingguan = [36, 30, 14, 25, 38, 16, 28, 12, 40]

hasil_analisis = analisis_suhu(suhu_mingguan, batas_panas=34, batas_dingin=16)

print(f"Hari Panas  (≥ {35}°C): {hasil_analisis['Panas']}")
print(f"Hari Normal (antara {15}°C - {35}°C): {hasil_analisis['Normal']}")
print(f"Hari Dingin (≤ {15}°C): {hasil_analisis['Dingin']}")
```
